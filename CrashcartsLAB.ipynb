```python
# Slow speed data and characteristics
slowData = [1.09,1.37,1.32,1.07,0.98,0.81]
slowMean = np.mean(slowData)
slowSigma = np.std(slowData)
slowVar = 100.*slowSigma/slowMean

fastdData = [1.7,1.67,1.76,1.65,1.68,1.62]
fastMean = np.mean(fastdData)
fastSigma = np.std(fastdData)
fastVar = 100.*fastSigma/fastMean

# Data check/print
print("Our average slow speed is %.3f m/s with a standard deviation of %.3f." % (slowMean,slowSigma))
print("The variability is %.2f%%" % slowVar)
print("Our average fast speed is %.3f m/s with a standard deviation of %.3f." % (fastMean,fastSigma))
print("The variability is %.2f%%" % fastVar)
```

```python
# range of t values -- choose lower and upper limits of range
modelTime = np.linspace(0.,10.,20)

# actual model parameters - slope and intercept
modelSlow = slowMean
modelSlowInt = 0

# generate x values from model -- exactly the equation in the cell above.
modelSlowX = modelSlowInt + modelSlow*modelTime
```

```python
fig1, ax1 = plt.subplots()

ax1.plot(modelTime, modelSlowX, color = 'green', linestyle = '-', linewidth = 2., label = " slow skateboard")

# a more explicit way to set labels
plt.xlabel('Time (Strange Quarks)', fontsize = 10)
plt.ylabel('Position (Charm Quarks)', fontsize = 10)
plt.title('Chanse\'s Edit', fontsize = 20)

ax1.grid()
fig1.set_size_inches(12, 8)
plt.legend(loc= 4)
plt.show()
```

***

```python
# width of the 'skateboard' front to back
cartLength = 0.5
frontOffset = cartLength/2.
backOffset = cartLength/2.

# generate location of front and back corners. Same slope different initial values
frontSlowX = (modelSlowInt + frontOffset) + modelSlow*modelTime
backSlowX = (modelSlowInt - backOffset) + modelSlow*modelTime
```

```python
fig2, ax2 = plt.subplots()

ax2.plot(modelTime, modelSlowX, color = 'green', linestyle = '-', linewidth = 2., label = " slow skateboard")
ax2.plot(modelTime, frontSlowX, color = 'red', linestyle = ':', linewidth = 1., label = "front")
ax2.plot(modelTime, backSlowX, color = 'red', linestyle = ':', linewidth = 1., label = "back")

# a more explicit way to set labels
plt.xlabel('time (Muon)', fontsize = 10)
plt.ylabel('position (Tau)', fontsize = 10)
plt.title('Chanses\'s Edit', fontsize = 20)

ax2.grid()
fig2.set_size_inches(12, 8)
plt.legend(loc= 4)
plt.show()

print("Question : Are you clear about why all the plotted lines are parallel?")
print("Yes, I understand that is the front and back of cart because the cart occupies space, which we need to account for.")
```

```python
# Slope and intercept fast cart
modelFast = fastMean
modelFastInt = 3.

# generate x values from model -- exactly the equation in the cell above.
modelFastX = modelFastInt + modelFast*modelTime

# faster front speed and slower back speed.
frontFastFast = fastMean + fastSigma # front corner speed for faster board
backFastSlow = fastMean - fastSigma # back corner speed for slower board

# generate location of front and back corners. different slopes per discussion
frontFFX = (modelFastInt + frontOffset) + frontFastFast*modelTime
backFSX = (modelFastInt - backOffset) + backFastSlow*modelTime

# plot it
fig4, ax4 = plt.subplots()

# slow skateboard
ax4.plot(modelTime, modelSlowX, color = 'purple', linestyle = '-', linewidth = 2., label = " slow skateboard")
ax4.plot(modelTime, frontSFX, color = 'red', linestyle = ':', linewidth = 1., label = "front + sigma")
ax4.plot(modelTime, backSSX, color = 'red', linestyle = ':', linewidth = 1., label = "back - sigma")

# fast skateboard
ax4.plot(modelTime, modelFastX, color = 'orange', linestyle = '-', linewidth = 2., label = " fast skateboard")
ax4.plot(modelTime, frontFFX, color = 'brown', linestyle = ':', linewidth = 1., label = "front + sigma")
ax4.plot(modelTime, backFSX, color = 'brown', linestyle = ':', linewidth = 1., label = "back - sigma")

# a more explicit way to set labels
plt.xlabel('Time (Seconds)', fontsize = 10)
plt.ylabel('Position (Meters)', fontsize = 10)
plt.title('Chanse\'s Edit', fontsize = 20)

# controlling x and y limits on plot to keep things where I want them
plt.xlim([0, 9])
plt.ylim([0, 15])

ax4.grid()
fig4.set_size_inches(12, 8)
plt.legend(loc= 4)
plt.show()
```

#### If you're going faster, the cart will pass the point of destination quicker, therefore front and back of sigma for the fast skateboard will take up less area.
#### If you're going slower, the 'skateboard' will be crossing the destination as a slower pace, therefore it will have a greater area because it spends more time at the destination.
#### Thinking about it from a quantum perspective those are the areas where the cart could be at that time.

```python
modelFast = fastMean
modelFastInt = 3
modelFastX = modelFastInt + modelFast*modelTime

# faster front speed and slower back speed.
frontFastFast = fastMean + fastSigma # front corner speed for faster board
backFastSlow = fastMean - fastSigma # back corner speed for slower board

# generate location of front and back corners. different slopes per discussion
frontFFX = (modelFastInt + frontOffset) + frontFastFast*modelTime
backFSX = (modelFastInt - backOffset) + backFastSlow*modelTime

# how long should fast skateboard wait?
waitTime = 2.82 # in s
hallLong = 9. # distance from start point in m
hallShort = 6.
hallDiff = hallLong - hallShort

# plot it
fig5, ax5 = plt.subplots()

# slow skateboard
ax5.plot(modelTime, modelSlowX, color = 'green', linestyle = '-', linewidth = 2., label = " slow skateboard")
ax5.plot(modelTime, frontSFX, color = 'red', linestyle = ':', linewidth = 1., label = "front + sigma")
ax5.plot(modelTime, backSSX, color = 'red', linestyle = ':', linewidth = 1., label = "back - sigma")

# fast skateboard
ax5.plot(modelTime + waitTime, modelFastX, color = 'blue', linestyle = '-', linewidth = 2., label = " fast skateboard")
ax5.plot(modelTime + waitTime, frontFFX, color = 'violet', linestyle = ':', linewidth = 1., label = "front + sigma")
ax5.plot(modelTime + waitTime, backFSX, color = 'violet', linestyle = ':', linewidth = 1., label = "back - sigma")

# draw a horizontal line at location of long hallway distance
ax5.hlines(hallLong, 0., 10, color = 'magenta', linestyle = '-', linewidth = 3.)

# a more explicit way to set labels
plt.xlabel('time (seconds)', fontsize = 10)
plt.ylabel('position (meters)', fontsize = 10)
plt.title('Bruce\'s Model', fontsize = 20)

# controlling x and y limits on plot to keep things where I want them
plt.xlim([0, 12])
plt.ylim([0, 12])

ax5.grid()
fig5.set_size_inches(12, 8)
plt.legend(loc=4)
plt.show()

print("Did the fast skateboard get to 9 m before or after the slow skateboard?")  

print("The fast skateboard reached 9 meters before the slow skateboard.")  

print("How far apart were they when one passed in front of the other?")  

print("The red lines indicate that the slow skateboard would have hit the back of the fast skateboard. The center of the fast skateboard passes 9 meters after 6.8 seconds and the center of the slow skateboard passes 9 meters at just about 8.1 seconds.")

print("If the fast skateboard only has 6 m to get to the intersection how do I plot that?")

print("You would want to add +3 for the Y intercept so the the skateboard only has to travel 6 m")  
```
